```python
import matplotlib.pyplot as plt
import numpy as np

# Create a grid of values for the gradient
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)

# Define the gradient function
Z = X + Y

# Plot the gradient using pcolormesh
plt.figure(figsize=(6, 6))
plt.pcolormesh(X, Y, Z, shading='auto', cmap='viridis')

# Set square aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

# Display the plot
plt.colorbar(label="Gradient Intensity")
plt.title("Gradient Square")
plt.show()
```

```python
def custom_heat_map(grayscale):
    MAX_SCALE = 255
    if grayscale<0:
        col = (0,0,255)
        
    elif grayscale<MAX_SCALE/4:
        col = (0,(grayscale)*4,255)
    
    elif grayscale<MAX_SCALE/2:
        col = (0,255,255-(grayscale-MAX_SCALE/4)*4)
    
    elif grayscale<3*MAX_SCALE/4:
        col = ((grayscale-MAX_SCALE/2)*4,255,0)
    
    elif grayscale<MAX_SCALE:
        col = (255,255-(grayscale-3*MAX_SCALE/4)*4,0)
        
    else:
        col = (255,0,0)
    
    return tuple(value/255 for value in col)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Define a custom colormap by providing a list of colors
col = (255, 1, 0)  # Red in RGB (0-255 scale)
white = (1,1,1)
low_col = (255,255,0)
low_col =  tuple( (col[i]*0.4 for i in range(3)))
colors = [col,low_col]
# colors = ['red','yellow']

# Create the custom colormap
cmap = LinearSegmentedColormap.from_list("custom_cmap", colors)

# Create a grid for displaying the colormap (a radial gradient)
x = np.linspace(-1, 1, 500)
y = np.linspace(-1, 1, 500)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
gradient = 1 - R
gradient = np.clip(gradient, 0, 1)


# Display the gradient using the custom colormap
plt.imshow(gradient, extent=(-1, 1, -1, 1), origin='lower', cmap=cmap)

# Add a title and colorbar
plt.title('Circular Gradient with Custom Colormap')
plt.colorbar(label='Gradient Intensity')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a grid of points
x = np.linspace(-1, 1, 500)
y = np.linspace(-1, 1, 500)
X, Y = np.meshgrid(x, y)

# Calculate the radial distance from the center
R = np.sqrt(X**2 + Y**2)

# Define the circular gradient (e.g., linear from 0 at center to 1 at the edges)
gradient = R  # Subtract R from 1 to have the intensity highest in the center

# Ensure the gradient is within the [0, 1] range
gradient = np.clip(gradient, 0, 1)

# Display the circular gradient using imshow
plt.imshow(gradient, extent=(-1, 1, -1, 1), origin='lower', cmap='autumn_r')

# Set aspect ratio to make the plot square
plt.gca().set_aspect('equal')

# Add a title and colorbar
plt.title('Circular Gradient')
plt.colorbar(label='Gradient Intensity')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

def display_two_colors(color1, color2):
    """
    Displays two colors side by side in a Matplotlib plot.

    Parameters:
    color1 (str or tuple): The first color (name or RGB tuple)
    color2 (str or tuple): The second color (name or RGB tuple)
    """

    cl1 = tuple(val/255 for val in color1)
    cl2 = tuple(val/255 for val in color2)
    color1 = cl1
    color2 = cl2
    # Create a figure with two subplots
    fig, ax = plt.subplots(1, 2, figsize=(6, 3))
    
    # Set the background color for each subplot
    ax[0].set_facecolor(color1)
    ax[0].set_title(f'Color 1: {color1}', fontsize=12)
    
    ax[1].set_facecolor(color2)
    ax[1].set_title(f'Color 2: {color2}', fontsize=12)
    
    # Remove axes for a clean look
    for a in ax:
        a.set_xticks([])
        a.set_yticks([])
    
    # Adjust layout and display the plot
    plt.tight_layout()
    plt.show()

# Example usage:
display_two_colors((255,0,0), (255,200,0))
```

```python
def adjust_brightness(color, factor):
    """
    Adjust the brightness of a color by a given factor.
    
    Args:
    - color (tuple): RGB color as (R, G, B)
    - factor (float): Factor to lighten (>1) or darken (<1). E.g., 1.2 for 20% lighter, 0.8 for 20% darker.
    
    Returns:
    - tuple: Adjusted RGB color.
    """
    r, g, b = color
    r = min(255, max(0, int(r * factor)))
    g = min(255, max(0, int(g * factor)))
    b = min(255, max(0, int(b * factor)))
    return (r, g, b)

# Example usage
original_color = (100, 150, 200)  # Some color
lightened_color = adjust_brightness(original_color, 1.5)  # Lighten by 20%
darkened_color = adjust_brightness(original_color, 0.8)   # Darken by 20%

print("Lightened:", lightened_color)
print("Darkened:", darkened_color)


display_two_colors(darkened_color, lightened_color)
```
